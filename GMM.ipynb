```python
import numpy as np
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt
```

```python
# Generate data
N1 = 50
N2 = 100
mean = np.array([0.0, 0.0])
cov = np.array([[.01, 0.0], [0.0, .01]])
x1, y1 = np.random.multivariate_normal(mean, cov, N1).T
angle = np.arange(-np.pi * 0.5, np.pi * 0.5, 0.05)
x2 = np.cos(angle) + np.random.normal(0, 0.01, angle.shape)
y2 = np.sin(angle)+ np.random.normal(0, 0.01, angle.shape)

plt.plot(x1, y1, 'x')
plt.plot(x2, y2, 'o')
plt.axis('equal')
plt.show()
```

```python
X1 = np.array([x1,y1]).T
X2 = np.array([x2,y2]).T
X = np.concatenate((X1,X2), axis = 0)
a,b = X.shape
gm = GaussianMixture(n_components=2, random_state=0).fit(X)
C = gm.predict(X)
for i in range(a):
    if C[i] == 0:
        plt.scatter(X[i][0], X[i][1], color='green')
    else:
        plt.scatter(X[i][0], X[i][1], color='red')
plt.show()
```
